```python

#페이지 네이션 - 기간 설정 - HTML에 페이지를 매기는 요청값이 들어가 있지 않음// 오류 발생
#마우스 클릭 이벤트 - 페이지 넘기기
#딕셔너리를 만들어서 데이터 표현 

    #URL추출 https://www.epeople.go.kr/cmmn/idea/redirect.do?ideaRegNo= 추출한 값
#https://www.epeople.go.kr/cmmn/idea/redirect.do?ideaRegNo=1AE-2404-0000479

#{}안에 추출한 게시물의 ideargno넣으면 해당 게시물의 url이 완성됩니다
# 
#https://www.epeople.go.kr/cmmn/idea/redirect.do?ideaRegNo={}
#<a href="#" class="go_detail" data-idearegno="1AE-2404-0000178" data-ideatycd="3A020010" data-ideastepcd="3A030010">
#<strong>2024년 진천읍 새마을지도자협의회 결식아동돕기 행사</strong>

#</a>
#a태그 안에 있는 idearegno 추출, URL폼안에 삽입하면, 해당 디테일로 연결되는 URL 완성
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import json

# Chrome 드라이버 설치
chrome_driver_path = ChromeDriverManager().install()


# 결과를 저장할 딕셔너리
result_dict = {}

# 페이지를 1부터 20까지 반복
for i in range(1, 21):
    # 페이지 URL을 형식화
    url = f"https://www.epeople.go.kr/nep/thk/subj/SubjThinkList.npaid?pageIndex={i}"

    # Chrome 웹 드라이버 초기화
    driver = webdriver.Chrome(service=Service(chrome_driver_path))
    
    # 페이지로 이동
    driver.get(url)
    
    # 요소가 로드될 때까지 대기
    driver.implicitly_wait(20)
    time.sleep(1)

    # 대화 상자 요소 찾기
    dialog_lst = driver.find_element(By.CLASS_NAME, "think_list.new").find_elements(By.CLASS_NAME, "thbox")

    # 페이지별 결과를 저장할 딕셔너리 초기화
    page_results = {}

    # 대화 상자 목록 반복
    for idx, dialog in enumerate(dialog_lst):
        dialog_data = {}

        # 제목 가져오기
        title = dialog.find_element(By.CLASS_NAME, "go_detail").text.strip()
        dialog_data['title'] = title

        # 기관 가져오기
        organization = dialog.find_element(By.CLASS_NAME, "thbot.basic.name").text.strip()
        dialog_data['organization'] = organization

        # 기간 가져오기
        periods = dialog.find_elements(By.CLASS_NAME, "thbot.basic")[1].text
        dialog_data['periods'] = periods

        # 참여자 수 가져오기
        participants = int(dialog.find_element(By.CLASS_NAME, "si.si01").text.strip("명 참여"))
        dialog_data['participants'] = participants

        # 추천 수 가져오기
        recommends = int(dialog.find_element(By.CLASS_NAME, "si.si04").text.strip())
        dialog_data['recommends'] = recommends

        # 비추천 수 가져오기
        no_recommends = int(dialog.find_element(By.CLASS_NAME, "si.si05").text.strip())
        dialog_data['no_recommends'] = no_recommends

        # a 태그 가져오기
        a_tag = dialog.find_element(By.CLASS_NAME, "go_detail")
        
        # a 태그가 있는 경우에만 처리
        if a_tag:
            # data-idearegno 속성값 가져오기
            data_idearegno = a_tag.get_attribute("data-idearegno")
            dialog_data['data_idearegno'] = data_idearegno

        # 결과 딕셔너리에 추가
        page_results[idx] = dialog_data

    # 페이지별 결과를 전체 결과 딕셔너리에 추가
    result_dict[f'page_{i}'] = page_results

    # 드라이버 종료
    driver.quit()

# JSON 파일로 저장
with open('resultdata(1-20).json', 'w') as json_file:
    json.dump(result_dict, json_file, indent=4)

'''
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# Chrome 드라이버 설치
chrome_driver_path = ChromeDriverManager().install()

#페이지 네이션 - 기간 설정 - HTML에 페이지를 매기는 요청값이 들어가 있지 않음// 오류 발생
#마우스 클릭 이벤트 - 페이지 넘기기
#딕셔너리를 만들어서 데이터 표현 

# 설치된 경로를 사용하여 Chrome 드라이버를 설정

for i in range(1, 20):
    # Format the URL with the page number
    url = f"https://www.epeople.go.kr/nep/thk/subj/SubjThinkList.npaid?page={i}"

    # Initialize the Chrome webdriver
    driver = webdriver.Chrome(service=Service(chrome_driver_path))
    
    # Get the webpage
    driver.get(url)
    
    # Wait for elements to load
    driver.implicitly_wait(20)
    time.sleep(1)
    '''
'''
with webdriver.Chrome(service=Service(chrome_driver_path)) as driver:
    driver.get("https://www.epeople.go.kr/nep/thk/subj/SubjThinkList.npaid")
    driver.implicitly_wait(30)
    time.sleep(3)
'''
'''
dialog_lst = driver.find_element(By.CLASS_NAME, "think_list.new").find_elements(By.CLASS_NAME, "thbox")
result_dict = {}
for idx, dialog in enumerate(dialog_lst):
    result_dict[idx] = {} #키 조정 필요
    box_titles = dialog.find_element(By.CLASS_NAME, "box_tit").find_elements(By.TAG_NAME, "span")

    box_title_lst = []
    for box_title in box_titles:
        box_title_lst.append(box_title.text.strip())
    result_dict[idx]['box_titles'] = box_title_lst

    title = dialog.find_element(By.CLASS_NAME, "go_detail").text.strip()
    result_dict[idx]['title'] = title

    organization = dialog.find_element(By.CLASS_NAME, "thbot.basic.name").text.strip()
    result_dict[idx]['organization'] = organization

    periods = dialog.find_elements(By.CLASS_NAME, "thbot.basic")[1].text
    result_dict[idx]['periods'] = periods

    participants = int(dialog.find_element(By.CLASS_NAME, "si.si01").text.strip("명 참여"))
    result_dict[idx]['participants'] = participants

    recommends = int(dialog.find_element(By.CLASS_NAME, "si.si04").text.strip())
    result_dict[idx]['recommends'] = recommends

    no_recommends = int(dialog.find_element(By.CLASS_NAME, "si.si05").text.strip())
    result_dict[idx]['no_recommends'] = no_recommends

    a_tag = dialog.find_element(By.CLASS_NAME, "go_detail")
    
    # a 태그가 존재하는 경우에만 처리
    if a_tag:
        # a 태그의 data-idearegno 속성값 가져오기
        data_idearegno = a_tag.get_attribute("data-idearegno")
        
        # 결과 딕셔너리에 data_idearegno 추가
        result_dict[idx]['data_idearegno'] = data_idearegno

'''
'''
    url_keys = dialog.find_element(By.CLASS_NAME, "go-detail").text.strip()
    data_idearegno = url_keys.get_attribute("data-idearegno")
    result_dict[idx]['data_idearegno'] = data_idearegno
    '''
'''    
result_dict

import json
with open("./crawling_data_CW24.json", "w", encoding='utf-8') as f:
    json.dump(result_dict, f, ensure_ascii=False, indent=4) 

    '''
            #추출해야하는 데이터 생각해보기 - 각자 연동을 해야함
        #1. 추천수를 기반으로 카운트 하여 보여줌
        #2. 대화 필터링은 어떻게 할거임? 
        #필터링한 데이터 - 주제, 참여도, 카테고리, 관할처, 기간, 추천, 비추천수, 아이디

        #데이터 베이스 스키마 구조 생각 (테이블 만들어 보기)
        # 스키마는 각 속성에 대한 레이블로 구성
```
